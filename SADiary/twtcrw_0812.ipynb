```python
text_query = '신난다 -RT'
count = 1000
tweet_list = []
try:
 # Creation of query method using parameters
    tweets = tweepy.Cursor(api.search,q=text_query).items(count)
 
 # Pulling information from tweets iterable object
#     tweets_list = [[tweet.created_at, tweet.id, tweet.text] for tweet in tweets]
    r = re.compile('#\S+|^https?\S+|@\S+')
    hangul = re.compile('[^ ㄱ-ㅣ가-힣|0-9|!$%&’()*+,.;<=>?\^_`{|}~-]+')
    for tweet in tweets:
        content = tweet.text
        content = hangul.sub('',content)
        content = re.sub(r, '', content)
        content = re.sub('  +','',content)
        if content != '' and content != ' ':
            tweet_list.append(content)
    
    exciting_tweets_df = pd.DataFrame(tweet_list, columns = ['raw_text'])
 
 # Creation of dataframe from tweets list
 # Add or remove columns as you remove tweet information
except BaseException as e:
    print('failed on_status,',str(e))

exciting_tweets_df
```

```python
exciting_tweets_df.to_csv("./exciting.csv", mode='w',index=True,header=True, encoding="utf-8-sig")
```
